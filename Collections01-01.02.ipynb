```python
class ContaCorrente:

  def __init__(self, codigo):
    self.codigo = codigo
    self.saldo = 0

  def deposita(self, valor):
    self.saldo += valor

  def __str__(self):
    return '[>> Codigo {} Saldo {} <<]'.format(self.codigo, self.saldo)
```

```python
conta_da_nala = ContaCorrente(15)
print(conta_da_nala)
```

```python
conta_da_nala.deposita(254)
print(conta_da_nala)
```

```python
conta_da_carol = ContaCorrente(6253)
conta_da_carol.deposita(10)
print(conta_da_carol)
```

```python
contas = [conta_da_nala, conta_da_carol]
```

```python
def deposita_para_todas(contas):
  for conta in contas:
    conta.deposita(100)

contas = [conta_da_nala, conta_da_carol]
print(contas[0], contas[1])
deposita_para_todas(contas)
print(contas[0], contas[1])
```

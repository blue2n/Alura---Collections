```python
idades = [12, 23, 34, 45, 56, 67, 12, 78]

for i in range(len(idades)):
  print(i, idades[i])
```

```python
range(len(idades)) #lazy
```

```python
enumerate(idades) #lazy
```

```python
list(range(len(idades))) #forçar geração de valores
```

```python
list(enumerate(idades))
```

```python
for valor in enumerate(idades):
  print(valor)
```

```python
usuarios = [('Nala', 2, 2021), ('Io', 22, 2001), ('Carol', 14, 2009)]

for nome, idade, ano in usuarios:  #maneira mais legivel, logo prazo
  print(nome)
```

```python
for nome, _, _ in usuarios:
  print(nome)
```
